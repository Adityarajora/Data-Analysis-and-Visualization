```python
l=[]
L=[]
s1=np.random.normal(loc=20,scale=5,size=1000)
s2=np.random.normal(loc=0,scale=1,size=1000)
s1.sort()
s2.sort()
for i in range(0,100):
    l.append(np.percentile(s1,i))
for i in range(0,100):
    L.append(np.percentile(s2,i))

plt.plot(L,l)
plt.show()
```
